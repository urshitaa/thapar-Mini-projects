Q9. Write to the file 100 random strings whose length between 10 and 15.

```python
import random 
import string 
with open("Q9.txt", "w") as f:  
    chars = string.ascii_letters
    for i in range(100):
        length = random.randrange(10,15)
        str = ''.join(random.choices(chars, k = length))
        print(str)
        f.write(str + "\n")
f.close()
```
