# Q1. Create a file that contains 1000 lines of random strings.

```python
import random as r
import string as s
```

```python
with open("random_strings.txt", "w") as file:
  for i in range(1000):
    stri = ""
    x = r.randint(11,100)
    for j in range(x):
        stri += r.choice(s.ascii_letters)
    print(stri)
    file.write(stri + "\n")

file.close()
```

# Q2. Create a file that contains multiple lines of random strings and file size must be 5 MB.

```python
import random as r
import string as s

with open("random_strings_5MB.txt", "w") as file:
  target_size = 5 * 1024 * 1024  # 5 MB in bytes
  written = 0

  while(written < target_size):
    line = ''.join(r.choices(s.ascii_letters, k = r.randint(31,100)))
    file.write(line + "\n")
    written += len(line.encode('utf-8'))
print("File Generated successfully")
```

# Q9. Convert all the videos of Q8 and convert it to audio. [Hint: Find the package from pypi.org and use it]

```python
import subprocess
import os

def convert_video_to_audio(video_path, output_audio_path=None):
    if not os.path.exists(video_path):
        print("Video file not found.")
        return

    # Set default output name if not provided
    if output_audio_path is None:
        base = os.path.splitext(video_path)[0]
        output_audio_path = base + '.mp3'

    try:
        subprocess.run([
            'ffmpeg',
            '-i', video_path,
            '-vn',                 # no video
            '-ab', '192k',         # audio bitrate
            '-ar', '44100',        # audio sample rate
            '-y',                  # overwrite output file if exists
            output_audio_path
        ], check=True)
        print(f"Converted successfully: {output_audio_path}")
    except subprocess.CalledProcessError:
        print("Error converting video to audio.")

# Example usage:
convert_video_to_audio("/content/videos/Machine Learning for Everybody  Full Course.mp4")
```

# Q10. Create an automated pipeline using multi-threading for:  Automatic Download of 100 Videos from YouTube  Convert it to Audio.

```python
import os
import subprocess
from concurrent.futures import ThreadPoolExecutor
import yt_dlp

# Folder setup
VIDEO_DIR = "videos"
AUDIO_DIR = "audios"
os.makedirs(VIDEO_DIR, exist_ok=True)
os.makedirs(AUDIO_DIR, exist_ok=True)

# Function to download one video using yt-dlp
def download_video(url):
    try:
        ydl_opts = {
            'format': 'mp4',
            'outtmpl': os.path.join(VIDEO_DIR, '%(title)s.%(ext)s'),
            'quiet': True
        }
        with yt_dlp.YoutubeDL(ydl_opts) as ydl:
            info = ydl.extract_info(url, download=True)
            return os.path.join(VIDEO_DIR, ydl.prepare_filename(info))
    except Exception as e:
        print(f"Failed to download {url}: {e}")
        return None

# Function to convert a video to MP3 using ffmpeg
def convert_to_audio(video_path):
    if video_path is None:
        return
    try:
        base = os.path.splitext(os.path.basename(video_path))[0]
        audio_path = os.path.join(AUDIO_DIR, f"{base}.mp3")
        subprocess.run([
            'ffmpeg', '-i', video_path,
            '-vn', '-ab', '192k', '-ar', '44100', '-y', audio_path
        ], stdout=subprocess.DEVNULL, stderr=subprocess.DEVNULL)
        print(f"Converted: {base}.mp3")
    except Exception as e:
        print(f"Failed to convert {video_path}: {e}")

# Function that wraps both stages
def process_video(url):
    video_path = download_video(url)
    convert_to_audio(video_path)

# Collect video URLs using a search (you can also manually supply 100 URLs)
def get_video_urls(search_term="Machine Learning", count=100):
    ydl_opts = {
        'quiet': True,
        'default_search': 'ytsearch100',
        'extract_flat': 'in_playlist',
        'force_generic_extractor': True
    }
    urls = []
    with yt_dlp.YoutubeDL(ydl_opts) as ydl:
        info = ydl.extract_info(f"ytsearch{count}:{search_term}", download=False)
        for entry in info['entries']:
            urls.append(entry['url'])
    return urls[:count]

# MAIN PIPELINE
if __name__ == "__main__":
    print("Fetching video URLs...")
    video_urls = get_video_urls("Machine Learning", 100)

    print("Starting download and conversion with multithreading...")
    with ThreadPoolExecutor(max_workers=5) as executor:
        executor.map(process_video, video_urls)

    print("All done.")
```

# Q10. Create an automated pipeline using multi-threading for: Automatic Download of 500 images of Dog from GoogleImages  Rescale it to 50%.

```python
from icrawler.builtin import GoogleImageCrawler
from PIL import Image
import os
from concurrent.futures import ThreadPoolExecutor

# Folder setup
IMG_DIR = "dog_images"
RESIZED_DIR = "dog_images_resized"
os.makedirs(IMG_DIR, exist_ok=True)
os.makedirs(RESIZED_DIR, exist_ok=True)

# STEP 1: Download Images from Google
def download_dog_images():
    crawler = GoogleImageCrawler(storage={'root_dir': IMG_DIR})
    crawler.crawl(keyword='dog', max_num=500)

# STEP 2: Resize image to 50%
def resize_image(image_file):
    try:
        img_path = os.path.join(IMG_DIR, image_file)
        img = Image.open(img_path)
        new_size = (img.width // 2, img.height // 2)
        img = img.resize(new_size)
        output_path = os.path.join(RESIZED_DIR, image_file)
        img.save(output_path)
        print(f"Resized: {image_file}")
    except Exception as e:
        print(f"Failed to resize {image_file}: {e}")

# MAIN
if __name__ == "__main__":
    print(" Downloading images...")
    download_dog_images()

    print(" Resizing images with multithreading...")
    image_files = [f for f in os.listdir(IMG_DIR) if f.lower().endswith(('.jpg', '.jpeg', '.png'))]

    with ThreadPoolExecutor(max_workers=8) as executor:
        executor.map(resize_image, image_files)
```
