```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df_train = pd.read_csv("/content/drive/MyDrive/ML & DL /Black Friday /train.csv")
df_test = pd.read_csv("/content/drive/MyDrive/ML & DL /Black Friday /test.csv")
df_test.head(5)
```

```python
df['Gender'] = df['Gender'].map({'F': 0, 'M': 1})
df.head()
```

```python
df['Age'] = df['Age'].map({
    '0-17':1,
    '18-25' :2,
    '26-35' :3,
    '36-45' :4,
    '46-50' :5,
    '51-55' :6,
    '55+' :7
})
df.head()
```
